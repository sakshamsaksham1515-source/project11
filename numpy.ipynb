```python
import numpy as np
import matplotlib.pyplot as plt

# Exam marks data
marks = [45, 52, 58, 62, 65, 68, 70, 71, 72, 74, 76, 78, 82, 90, 96]

# Calculate mean and standard deviation
mean = np.mean(marks)
std_dev = np.std(marks)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Create histogram
plt.hist(marks, bins=6)

# Labels and title
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Normal Distribution of Exam Marks")

# Show graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

marks = [45, 52, 58, 62, 65, 68, 70, 71, 72, 74, 76, 78, 82, 90, 96]

# Mean and standard deviation
mean = np.mean(marks)
std = np.std(marks)

print("Mean:", mean)
print("Standard Deviation:", std)

# Histogram
plt.hist(marks, bins=6)

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Gaussian Distribution")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
marks = [55, 60, 65, 70, 75, 80, 85]

# Mean
mean = np.mean(marks)

# Standard deviation
std_dev = np.std(marks)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Histogram
plt.hist(marks, bins=5)

# Mean line
plt.axvline(mean)

# Labels
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Standard Deviation Example")

plt.show()
```
