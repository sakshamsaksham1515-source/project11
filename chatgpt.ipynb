```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Aman", "Ravi", "Sita", "Neha"],
    "Marks": [80, 70, 90, 85]
}

df = pd.DataFrame(data)

plt.bar(df["Name"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()
```
